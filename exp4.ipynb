```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("sentiment analysis.csv")
```

```python
df
```

```python
x_train,x_test,y_train,y_test=train_test_split(df['sentence'],df['sentiment'],test_size=0.3,random_state=42)
```

```python
vectorizer=CountVectorizer()
x_train_vector=vectorizer.fit_transform(x_train)
x_test_vector=vectorizer.transform(x_test)
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train_vector,y_train)
```

```python
y_pred=model.predict(x_test_vector)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy of sentiment analysis model:",round(accuracy*100,2),"%")
```

```python
sample=['I really enjoyed the movie']
sample_vector=vectorizer.transform(sample)
prediction=model.predict(sample_vector)
```
